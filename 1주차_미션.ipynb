```python
import pandas as pd

# 2011년 이후의 데이터만 필터링
filtered_df = df[df['Year'] >= 2011]

# 피벗 테이블 생성
pivot_df = filtered_df.pivot_table(index='Year', columns='Country', values='Life_Expectancy', aggfunc='mean')

# 인덱스를 제거하고 Year 열을 첫 번째 열로 변환
pivot_df.reset_index(inplace=True)

# 'Country' 인덱스 제거
pivot_df.columns.name = None

# 결과 출력
print(pivot_df.to_string(index=False))

```
